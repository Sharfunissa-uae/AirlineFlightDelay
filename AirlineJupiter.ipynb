```python
import matplotlib.pyplot as plt
df=pd.read_csv('/Users/nazeerahamed/Desktop/Project/airlinedelaycsv.csv')
x=df["Airline"]
y=df["DelayMinutes"]
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()
```

```python
x=df["Airline"]=="emirates"
y=df["DelayMinutes"]
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()
```

```python
categories=df["Airline"]
values=df["DelayMinutes"]

plt.bar(categories,values,color='blue',edgecolor='red',alpha=1)
plt.bar(categories,values,color='cyan',width=0.5)
plt.xlabel("Airline Names")
plt.ylabel("DelayMinutes")
plt.title("Bar Chart")
plt.show()
```

```python
x=df["Airline"]
y=df["DelayMinutes"]
plt.scatter(x,y,color='green',marker='D')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```
